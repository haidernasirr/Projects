```python
x1=np.array([20,1,98,9])
x2=np.array([0,1,0,9])
y1=np.array([96,81,8,6])
y2=np.array([6,8,87,6])
plt.plot(x1,x2,y1,y2)
plt.xlabel('Artificial intelligence')
plt.ylabel('Machine learning')
plt.title('AI&ML')
```

```python
x1=np.array([20,1,98,9])
x2=np.array([0,1,0,9])
y1=np.array([96,81,8,6])
y2=np.array([6,8,87,6])
font1={'family':'arial','color':'blue','size':20}
font2={'family':'serif','color':'red','size':40}
plt.plot(x1,x2,y1,y2)
plt.xlabel('Artificial intelligence',fontdict=font1)
plt.ylabel('Machine learning',fontdict=font1)
plt.title('AI&ML',fontdict=font2)
```
