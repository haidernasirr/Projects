```python
sns.scatterplot(x='total_bill',y='tip', data=tips)
plt.title('scatter plot of total bill vs tip')
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=20,kde=True)
plt.title('histogram')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips)
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

```python
sns.boxplot(x='day',y='tip', data=tips)
plt.title('box plot of day')
```

```python
sns.pairplot(tips)
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip',color='red', data=tips)
plt.title('customised scatterplot for bill vs tip')
```

```python
plt.Figure(figsize=(10,6))
sns.violinplot(x='day', y="total_bill", hue='time', data=tips ,split=False)

plt.title('customised Violinplot for bill, Tip, Time')
plt.xlabel('day')
plt.ylabel('time')
```

```python
plt.Figure(figsize=(10,6))
sns.barplot(x='day', y="tip", hue='sex', data=tips ,palette='pastel',errorbar=None)

plt.title('Average Tip amount by sex and day')
plt.xlabel('day')
plt.ylabel('average tip')
```

```python
plt.Figure(figsize=(10,6))
sns.countplot(x='day', hue='smoker', data=tips ,palette='pastel')

plt.title('Count of dinners by day and smoking preference')
plt.xlabel('day of the week')
plt.ylabel('count of dinners')

plt.legend(title='smoker')
```
